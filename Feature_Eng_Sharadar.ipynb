# In this notebook we will convert the Sharadar symbol names to Norgate names by using our symbol reference dictionary. 

```python
import numpy as np
import pandas as pd
import norgatedata 
import datetime
import pickle
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
path = 'D:/Udacity/final/'
```

import roic and ebitev

```python
russell_roic = pd.read_csv(path + 'russell_roic_1.csv',parse_dates=True,index_col = 0)
russell_ebitev = pd.read_csv(path + 'russell_ebitev_1.csv',parse_dates=True,index_col = 0)

```

import our symbol name dictionary

```python
funda_sym_ref = pickle.load(open(path + 'funda_sym_ref_1.pkl','rb'))
```

```python
funda_sym_ref
```

invert the key values

```python
funda_sym_ref_invert = dict([(value, key) for key, value in funda_sym_ref.items()]) 
```

change symbol names to Norgate's format

```python
russell_roic_pre = russell_roic.rename(funda_sym_ref_invert,axis=1)

russell_ebitev_pre = russell_ebitev.rename(funda_sym_ref_invert,axis=1)
```

```python
russell_roic_pre
```

export fundamentals with changed names and only up till 2020-09-29

```python
russell_roic_pre.iloc[:-3].ffill().to_csv(path + 'russell_roic_pre_f.csv')

russell_ebitev_pre.iloc[:-3].ffill().to_csv(path + 'russell_ebitev_pre_f.csv')
```
